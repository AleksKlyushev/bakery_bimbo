```python
city['State'] = city.State.replace('MÉXICO, D.F.','MÉXICO')
city['Town_HZ'] = [a[:4] for a in city.Town]
city['Town_HZ2'] = [a[5:].replace('AG.','').strip() for a in city.Town]
city['Adr_str'] = city.Town_HZ2 + ', ' + city.State +', MÉXICO'
city['lat'] = 0.0
city['lng'] = 0.0
city
```

```python
zz = city.groupby('Town_HZ2',as_index = False).agg({"Town_HZ": pd.Series.nunique})
zz[zz.Town_HZ >= 2]
```

```python
for t in range(len(city)):
    # Geocoding an address
    geocode_result = gmaps.geocode(city.Adr_str[t])
    if geocode_result != []:
        city.lat[t] = geocode_result[0]['geometry']['location']['lat']
        city.lng[t] = geocode_result[0]['geometry']['location']['lng']
```
